```python
from langchain_community.tools import tool

# defining a tool

@tool
def cube(num: int) -> int:
  """calculate cube of the input"""
  return num **3
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata
from langchain_core.messages import HumanMessage


model = ChatGoogleGenerativeAI(
    model= "gemini-2.5-flash",
    api_key = userdata.get("GEMINI_API_KEY")
)

# tool Binding
model_with_tools = model.bind_tools([cube])
```

```python
chat_history = []

chat_history.append(HumanMessage("calculate the cube of 10"))

# tool calling
result = model_with_tools.invoke(chat_history)
chat_history.append(result)

# tool Execution
chat_history.append(cube.invoke(result.tool_calls[0]))

model_with_tools.invoke(chat_history).content
```
